**Data Visualization**

```
#checking Null Values
df.isnull()
```

**Train Test Split**

```
#To spilt dependent and independent variables
X=df.drop('Scores',axis=1)
y=df.drop('Hours',axis=1)
```

```
#Print number of rows and columns
print('shape of X:',X.shape)
print('shape of y:',y.shape)
```

```
#split tha data into training set (80%) and a testing set (20%)
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size = 0.2, random_state=4)
```

```
X_test
```

```
y_test
```

```
#Print shape of train_test_
print("shape of X_train:",X_train)
print("shape of X_test:",X_test)
print("shape of y_train:",y_train)
print("shape of y_test:",X_test)
```

**Train The Model**

```
#To perform Linear Regression
from sklearn.linear_model import LinearRegression
```

```
my_model = LinearRegression()

result = my_model.fit(X_train, y_train)
```

**To Test Model**

```
predictions = result.predict(X_test)

predictions
```
